```python
import pandas as pd
import numpy as np
```

```python
# valores de utm de enero de cada año obtenidas de https://si3.bcentral.cl/Bdemovil/BDE/Series/MOV_SC_PR12
valor_utm = {2023: 61769,
             2022: 54442,
             2021: 50978,
             2020: 49673,
             2019: 48353,
             2018: 47019,
             2017: 46229,
             2016: 44955,
             2015: 43198,
             2014: 40935
            }
```

```python
year = 2014
df = pd.read_csv(f"Datos/03_primary/{year}.csv")
```

```python
limite_excento = 225*valor_utm[year]
```

```python
df["Excento_Aseo"] = np.where(df["Avaluo_Fiscal"] <= limite_excento, 1, 0)
```

```python
df.head(5)
```

```python
logic = (df["Semestre"]==1) #& (df["Excento_Aseo"]==1)
df.loc[logic].groupby(["Region", "Comuna", "Excento_Aseo"]).agg({"Excento_Aseo":"size", "Aseo":"sum"}).unstack()
```
